### What does Colab do?

Now that you have an idea of what notebooks are, lets talk about Google Colab interface. Google Colab is able to run python notebooks. It also has many helpful features exploration, organization, and validation.

#### Document Outline

Colab will show you the document outline. Click on the menu button on the left side of the screen to open and close the document outline.

#### Variable Explorer

Colab will show you all python objects that you have put in memory by running code cells. Click on the `{x}` button on the left side of the screen to see all the python objects and their values. You can hover or click on the Name to see more information about each object.

Notice that the objects do not get cleared between running code cells. The objects in memory stay in memory cell to cell!

#### File Browser

Colab will let you access files from a number of different locations.
```python
from matplotlib import pyplot as plt
import numpy as np

# Generate 100 random data points along 3 dimensions
x, y, scale = np.random.randn(3, 100)
fig, ax = plt.subplots()

# Map each onto a scatterplot we'll create with Matplotlib
ax.scatter(x=x, y=y, c=scale, s=np.abs(scale)*500)
ax.set(title="Some random data, created with pyplot")
plt.show()

# source: https://jupyter.org/try-jupyter/notebooks/?path=notebooks/Intro.ipynb
```
